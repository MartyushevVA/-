```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('https://raw.githubusercontent.com/Adelaaas/Data_science_basic_22-23_1/main/class_work_6/train_var_3.csv')
```

1

```python
fig, ax = plt.subplots(figsize=(10, 8))
x = df['MinTemp']
y = df['MaxTemp']
plt.xlabel("MinTemp", fontsize=14)  
plt.ylabel("MaxTemp", fontsize=14)
plt.title("Разброс минимальных и максимальных суточных температур", fontsize=22)
plt.grid()
plt.minorticks_on()
ax.scatter(x, y, c = 'deeppink', s = 0.5)
plt.show()
```

Различия максимальных и минимальных температур в разных городах приблизительно равны.

2

```python
plt.figure(figsize=(17,7))
cmap = sns.cubehelix_palette(as_cmap=True, light=.9)
sns.heatmap(df.isna().transpose(), cmap=cmap, cbar_kws={'label': 'Пропущенные значения'})
plt.xlabel("Индексы", fontsize=14)  
plt.ylabel("Критерии", fontsize=14)
plt.title("Количество пропусков в данных")
plt.grid()
plt.show()
```

Видно, что большая часть пропусков данных сконцентрированна в столбцах "Evaporation", "Sunshine", "Cloud9am", "Cloud3pm".

3

```python
fig, axes = plt.subplots(6, 3, figsize=(40, 20))

sns.boxplot(df['MinTemp'], ax=axes[0][0], width=0.3)
sns.boxplot(df['MaxTemp'], ax=axes[0][1], width=0.3)
sns.boxplot(df['Rainfall'], ax=axes[0][2], width=0.3)
sns.boxplot(df['Evaporation'], ax=axes[1][0], width=0.3)
sns.boxplot(df['Sunshine'], ax=axes[1][1], width=0.3)
sns.boxplot(df['WindGustSpeed'], ax=axes[1][2], width=0.3)
sns.boxplot(df['WindSpeed9am'], ax=axes[2][0], width=0.3)
sns.boxplot(df['WindSpeed3pm'], ax=axes[2][1], width=0.3)
sns.boxplot(df['Humidity9am'], ax=axes[2][2], width=0.3)
sns.boxplot(df['Humidity3pm'], ax=axes[3][0], width=0.3)
sns.boxplot(df['Pressure9am'], ax=axes[3][1], width=0.3)
sns.boxplot(df['Pressure3pm'], ax=axes[3][2], width=0.3)
sns.boxplot(df['Cloud9am'], ax=axes[4][0], width=0.3)
sns.boxplot(df['Cloud3pm'], ax=axes[4][1], width=0.3)
sns.boxplot(df['Temp9am'], ax=axes[4][2], width=0.3)
sns.boxplot(df['Temp3pm'], ax=axes[5][0], width=0.3)
sns.boxplot(df['RainToday'], ax=axes[5][1], width=0.3)
sns.boxplot(df['RainTomorrow'], ax=axes[5][2], width=0.3)
plt.show()
```

4

```python
sns.barplot(data = df, x = 'Location', y = 'RainTomorrow')
plt.xlabel("Город", fontsize=14)  
plt.ylabel("Количество дождливых дней", fontsize=14)
plt.title("Количество дождливых дней в городах")
plt.grid()
plt.show()
```

5

```python
plt.figure(figsize=(12,10), dpi= 80)
sns.heatmap(df.corr(), xticklabels=df.corr().columns, yticklabels=df.corr().columns, cmap='RdYlGn', center=0, annot=True)
plt.title('Корреляции в характеристиках', fontsize=20)
plt.show()
```
